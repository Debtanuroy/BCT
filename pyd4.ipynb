```python
def countfact(num):
    c=0
    for i in range(1, num+1):
        if(num%i==0):
            c+=1
    return c

def fibonacci(num):
    a=0
    b=1
    for _ in range (num):
        print(a," ",end=" ")
        a, b = b, a+b

def factorial(num):
    fact=1
    for i in range (1, num+1):
        fact*=i
    return fact

while True:
    print("\nChoose an operation:")
    print("1. Prime Number Check")
    print("2. Fibonacci Series")
    print("3. Factorial of a Number")
    print("4. Exit")

    choice = int(input("Enter your choice: "))

    if choice == 1:
        num=int(input("Enter the number to be checked: "))
        c=countfact(num)
        if c == 2:
            print("Prime")
        else:
            print("Not prime")
    elif choice == 2:
        num=int(input("Enter the number of terms: "))
        fibonacci(num)
    elif choice == 3:
        num=int(input("Enter a number: "))
        print("Factorial of ",num,": ", factorial(num))
    elif choice == 4:
        print("Exiting program...")
        break
    else:
        print("Invalid choice! Please try again.")
```

```python
import getpass

# User data storage
users = []

def sign_up():
    name = input("Enter your name: ")
    email = input("Enter your email: ")
    
    # Check if email is already registered
    if any(user['Email'] == email for user in users):
        print("Email already registered. Try logging in.")
        return
    
    password = getpass.getpass("Enter your password: ")
    confirm_password = getpass.getpass("Confirm your password: ")
    
    if password != confirm_password:
        print("Passwords do not match. Please try again.")
        return
    
    users.append({"Name": name, "Email": email, "Password": password})
    print("Sign-up successful!")

def login():
    email = input("Enter your email: ")
    password = getpass.getpass("Enter your password: ")
    
    for user in users:
        if user['Email'] == email and user['Password'] == password:
            print("Login successful!")
            return
    print("Invalid credentials. Try again.")

def display_users():
    if not users:
        print("No users registered yet.")
        return
    
    print("\nRegistered Users:")
    print("--------------------------------")
    print("{:<20} {:<30}".format("Name", "Email"))
    print("--------------------------------")
    for user in users:
        print("{:<20} {:<30}".format(user['Name'], user['Email']))
    print("--------------------------------")

def delete_user():
    email = input("Enter the email of the user to delete: ")
    global users
    
    if any(user['Email'] == email for user in users):
        users = [user for user in users if user['Email'] != email]
        print(f"User with email {email} has been deleted.")
    else:
        print("No user found with this email.")

def main():
    while True:
        print("\n1. Sign Up\n2. Login\n3. Display Users\n4. Delete User\n5. Exit")
        choice = input("Choose an option: ")
        
        if choice == "1":
            sign_up()
        elif choice == "2":
            login()
        elif choice == "3":
            display_users()
        elif choice == "4":
            delete_user()
        elif choice == "5":
            print("Exiting...")
            break
        else:
            print("Invalid choice. Try again.")

if __name__ == "__main__":
    main()
```
